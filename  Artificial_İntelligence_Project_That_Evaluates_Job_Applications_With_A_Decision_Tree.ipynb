 # Decison Tree Classification

In this project, we will lay the foundation for the artificial intelligence project that evaluates job applications by classifying with Decision Trees.

```python
import numpy as np
import pandas as pd
from sklearn import tree

df = pd.read_csv("DecisionTreesClassificationDataSet.csv")
```

```python
df.head()
```

Since the data in this project is a Turkish source, there are Turkish values in the codes.

The scikit-learn library expects everything to be numerical for decision trees to function properly. Therefore, we convert all 'Y' and 'N' values in our dataset to 0 and 1, respectively. Similarly, we update the education level to 'BS:0', 'MS:1', and 'PhD:2'. Using the map() function, cells with missing values or invalid entries will be filled with NaN.

```python
df
```

```python
duzetme_mapping = {'Y': 1,'N': 0}

df['IseAlindi'] = df['IseAlindi'].map(duzetme_mapping)
df['SuanCalisiyor?'] = df['SuanCalisiyor?'].map(duzetme_mapping)
df['Top10 Universite?'] = df['Top10 Universite?'].map(duzetme_mapping)
df['StajBizdeYaptimi?'] = df['StajBizdeYaptimi?'].map(duzetme_mapping)
duzetme_mapping_egitim = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Egitim Seviyesi'] = df['Egitim Seviyesi'].map(duzetme_mapping_egitim)
df.head()
```

```python
y = df['IseAlindi']
x = df.drop(['IseAlindi'],axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)



```

```python
#Let's make a prediction now:
# Someone with 5 years of experience, currently employed, and having worked at 3 previous companies, with an education level of Bachelor's.
# Not a graduate of a top-tier school

print(clf.predict([[5,1,3,0,0,0]]))
```

```python
# A person with a total of 2 years of work experience, having changed jobs 7 times, and being a graduate of a very good school, is currently not employed.
print (clf.predict([[2, 0, 7, 0, 1, 0]]))
```

```python
# A person with a total of 2 years of work experience, having changed jobs 7 times, and not being a graduate of a very good school, is currently employed.
print (clf.predict([[2, 1, 7, 0, 0, 0]]))
```

```python
# Someone with a total of 20 years of work experience, having changed jobs 5 times, and being a graduate of a good school, is currently not employed.
print (clf.predict([[20, 0, 5, 1, 1, 1]]))
```
